# Images processing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("images/plane.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
image_bgr = cv2.imread("images/plane.jpg", cv2.IMREAD_COLOR)  # BGR - Blue Green Red
image_bgr[0, 0]
```

```python
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
plt.show(image_rgb)
plt.axis("off")
plt.show()
```
